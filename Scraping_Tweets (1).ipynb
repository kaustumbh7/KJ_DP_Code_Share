```python
def remove_eng_char(text):
    alphabets = list('qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM():@1234567890')
    text = str(text)
    st2 = text
    for i in text:
        if i in alphabets:
            st2 = st2.replace(i,"")
    return(st2)
```

```python
for i in range(len(df_res['Hate'])):
    df_res['Hate'][i] = remove_eng_char(df_res['Hate'][i])
for i in range(len(df_res['Racial Abuse'])):
    df_res['Racial Abuse'][i] = remove_eng_char(df_res['Racial Abuse'][i])
for i in range(len(df_res['Obscenity'])):
    df_res['Obscenity'][i] = remove_eng_char(df_res['Obscenity'][i])
```
